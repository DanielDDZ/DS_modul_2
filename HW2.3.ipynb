# Агрегування даних та з'єднання таблиць

Для початку давайте подивимося на максимальну та мінімальну ціни для кожного з жанрів (використовуйте функції groupby та agg, для підрахунку мінімальних та максимальних значень використовуйте max та min). Не беріть усі стовпці, виберіть тільки потрібні вам

```python
df.groupby('genre').agg({'price' : ['min', 'max']})
```

Максимальна ціна для жанру Fiction: **Відповідь: 82**

Мінімальна ціна для жанру Fiction: **Відповідь: 0**

Максимальна ціна для жанру Non Fiction: **Відповідь: 105**

Тепер створіть новий датафрейм, який вміщатиме кількість книг для кожного з авторів (використовуйте функції groupby та agg, для підрахунку кількості використовуйте count). Не беріть усі стовпці, виберете тільки потрібні

```python
authors = df.groupby('author').agg(books_count = ('name', 'count'))
authors
```

Якої розмірності вийшла таблиця? **Відповідь: (248, 1)**

```python
authors.shape
```

Який автор має найбільше книг? **Відповідь: Jeff Kinney**

```python
max_books = authors["books_count"].max()
authors.query('books_count == @max_books')
```

```python
authors.sort_values(by='books_count').tail(1)
```

Скільки книг цього автора? **Відповідь: 12**

Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean). Не беріть усі стовпці, виберете тільки потрібні

```python
authors_reting = df.groupby('author').agg(average_rating = ('user_rating', 'mean'))
authors_reting
```

У якого автора середній рейтинг мінімальний? Відповідь:

```python
min_rating = authors_reting["average_rating"].min()
authors_reting.query('average_rating == @min_rating')
```

```python
authors_reting.sort_values(by='average_rating').head(1)
```

Який у цього автора середній рейтинг? **Відповідь: 3.9**

З'єднайте останні два датафрейми так, щоб для кожного автора було видно кількість книг та середній рейтинг (Використовуйте функцію concat з параметром axis=1). Збережіть результат у змінну

```python
result_authors = pd.concat([authors, authors_reting], axis=1)
result_authors
```

Відсортуйте датафрейм за зростаючою кількістю книг та зростаючим рейтингом (використовуйте функцію sort_values)

```python
result_authors.sort_values(['books_count', 'average_rating'])
```

Який автор перший у списку? **Відповідь: Muriel Barbery**

```python
result_authors.sort_values(['books_count', 'average_rating']).head(1)
```

# Візуалізація

```python
books = df['genre'].value_counts()
explode = [0.03, 0.03]

colors = ['y', 'b']
plt.pie(
    books.values,
    labels=books.index,
    shadow=True,
    colors=colors,
    explode=explode,
    autopct="%1.1f%%",
    pctdistance=0.8,
    labeldistance=1.1,
    startangle=90
)

plt.title("Розподіл за жанром книг", fontweight="bold", fontsize=14)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(x='price', y='user_rating', hue='year', data=df)
plt.title('Залежність рейтингу від ціни з 2010 по 2018 рік', fontsize=15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.stripplot(x='year', y='reviews', hue='user_rating', data=df)
plt.title('Залежність рейтингу від кількісті відгуків з 2009 по 2019 рік', fontsize=15)
plt.show()
```
