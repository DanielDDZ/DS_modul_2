# Частина перша: Знайомство з Pandas.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_html("https://uk.m.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C",
                  match="Регіон", decimal=',', thousands='.')
df = df[0]
df
```

Вивести перші рядки таблиці за допомогою методу head

```python
df.head(4)
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
df.shape
```

Замініть у таблиці значення "—" на значення NaN

```python
df = df.replace({"—": np.nan})
df
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
df.dtypes
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df['2014'] = pd.to_numeric(df['2014'])
df['2019'] = pd.to_numeric(df['2019'])
df.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
df.isnull().sum()
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
last_row = len(df) - 1
df.drop(axis=0, index=last_row, inplace=True)
df
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df = df.fillna({"1950": df['1950'].mean(), "1960": df['1960'].mean(), "1970": df['1970'].mean(), "1990": df['1990'].mean(),
                "2000": df['2000'].mean(), "2012": df['2012'].mean(), "2014": df['2014'].mean(), "2019": df['2019'].mean()})
df
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
df[df['2019'] > df['2019'].mean()][["Регіон", "2019"]].sort_values(by="2019", ascending=False)
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
df.sort_values(by="2014", ascending=False).head(1)[["Регіон", "2014"]]
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
fig, ax = plt.subplots(figsize=(10, 4))
df.plot(x="Регіон", y="2019", kind="bar",
        fontsize=14, color='y', ax=ax, grid=True)
plt.title("Діаграма народжуваності по регіонах у 2019 році", fontdict = {'fontsize' : 20})
ax.set_xlabel("Регіон", size=12, color="b")
ax.set_ylabel("Коефіцієнт народжуваності",size=12, color="b")
plt.show()
```

# Візуалізація

```python
fig, ax = plt.subplots(figsize=(10, 6))
date = df.loc[df['Регіон'] == 'Одеська', '1950':'2019'].columns.tolist()
k_per_year = df.loc[df['Регіон'] == 'Одеська', '1950':'2019'].values[0]
plt.plot(date, k_per_year, label='Коефіцієнт народжуваності', linestyle="-", marker="o", color="y")
plt.xlabel('Дата', fontsize=15, color='b')
plt.ylabel('Коефіцієнт народжуваності', fontsize=15, color='b')
plt.title('Коефіцієнт народжуваності в Одеській області', fontsize=15)
plt.xlim(0, 7)
plt.grid()
plt.show()
```

```python
sort_df = df.sort_values(['2019'], ascending=False)
fig, ax = plt.subplots(figsize=(10, 6))
sns.pointplot(x="2019", y="Регіон", data=sort_df, color='b')
plt.title('Коефіцієнт народжуваності в 2019 по обл', fontsize=15)
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(10,6))
axs.plot(df['Регіон'], df['2019'], label='2019', linestyle="--", color="y")
axs.plot(df['Регіон'], df['2014'], label='2014', linestyle=":", color="b")

plt.xlabel('Регіон', fontsize='small', color="y")
plt.ylabel('Народжуваність', fontsize='small', color="y")
plt.title('Народжуваність по регіонах України')
plt.xticks(rotation=90)
plt.grid()
plt.legend()
plt.show()
```
