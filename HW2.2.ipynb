Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python

Використовуючи метод groupby, виконайте групування за стовпчиком "Посада"

```python
group_position = python_data.groupby(["Должность"])
group_position
```

Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"

```python
salary_group_data = group_position['Зарплата.в.месяц'].agg(['min', 'max'])
salary_group_data
```

Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"

```python
def fill_avg_salary(data):
  return np.average([data['min'], data['max']])

salary_group_data['avg'] = salary_group_data.apply(fill_avg_salary, axis=1)
salary_group_data
```

Створіть описову статистику за допомогою методу describe для нового стовпчика.

```python
salary_group_data['avg'].describe()
```

Збережіть отриману таблицю в CSV файл

```python
salary_group_data.to_csv("Salary_per_month.csv")
```

# Візуалізація

```python
languages = df['Язык.программирования'].value_counts(normalize=True)
other_languages = languages[languages < 0.03].sum()
languages = languages[languages >= 0.03]
languages['Інші'] = other_languages

fig, ax = plt.subplots(figsize=(10, 6))
explode = [0.15, 0.15, 0.15, 0, 0, 0, 0, 0, 0]
#ax.pie(languages.values, labels=languages.index, autopct='%1.1f%%', explode=explode)
plt.pie(
    languages.values,
    labels=languages.index,
    shadow=True,
    explode=explode,
    autopct="%1.1f%%",
    pctdistance=0.80,
    labeldistance=1.15,
)
plt.title('Мови програмування за популярністю', color='b')
plt.show()
```

```python
grouped = df.groupby('Язык.программирования')['Зарплата.в.месяц'].mean().reset_index()
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(grouped['Язык.программирования'], grouped['Зарплата.в.месяц'], color='y')
ax.set_xlabel('Мова програмування', fontsize=15, color='blue')
ax.set_xticklabels(grouped['Язык.программирования'], rotation=45, ha='right')
ax.set_ylabel('Середня зарплата на місяць', fontsize=15, color='blue')
ax.set_title('Залежність середньої заробітньої платні від мови програмуівання', fontsize=15)
plt.show()
```

```python

data = salary_group_data.groupby('Должность')['avg'].mean().reset_index()

plt.bar(data['Должность'], data['avg'], color=["b", "y", "r", "g", "m"])
plt.xlabel("Position", fontsize=10, color="b")
plt.ylabel("Salary range", fontsize=10, color="b")
plt.title("Average salary per position in 2017", fontsize=15)

plt.xticks(rotation=90)

plt.show()
```
